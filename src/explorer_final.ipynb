```python
import os

directory = "../data"
database_path = os.path.join(directory, "movies_database.db")
csv_path = os.path.join(directory, "/processed/clean_data.csv")
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
import sqlite3

conn = sqlite3.connect(database_path)

movies.to_sql("movies_table", conn, if_exists = "replace", index = False)
credits.to_sql("credits_table", conn, if_exists = "replace", index = False)
```

```python
query ='''select * from movies_table inner join credits_table on 
movies_table.title = credits_table.title
'''
```

```python
total_data = pd.read_sql_query(query, conn)
conn.close()
```

```python
total_data.head(2)
```

```python
total_data.columns
```

```python
total_data_new = total_data[['movie_id', 'title', 'overview', 'genres', 'keywords','cast','crew']]
```

```python
total_data_new.head(2)
```

```python
total_data_new = total_data_new.loc[:, ~total_data_new.T.duplicated()]
```

```python
total_data_new = total_data_new[['movie_id', 'title', 'overview', 'genres', 'keywords','cast','crew']]
```

```python
total_data_new.head(2)
```

```python
import json

def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value
    
total_data_new["genres"] = total_data_new["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
total_data_new["keywords"] = total_data_new["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
```

```python
total_data_new["cast"] = total_data_new["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)

total_data_new["crew"] = total_data_new["crew"].apply(lambda x: " ".join([crew_member['name'] for crew_member in load_json_safe(x) if crew_member['job'] == 'Director']))

total_data_new["overview"] = total_data_new["overview"].apply(lambda x: [x])

total_data_new.head()
```

```python
total_data_new["overview"] = total_data_new["overview"].apply(lambda x: [str(x)])
total_data_new["genres"] = total_data_new["genres"].apply(lambda x: [str(genre) for genre in x])
total_data_new["keywords"] = total_data_new["keywords"].apply(lambda x: [str(keyword) for keyword in x])
total_data_new["cast"] = total_data_new["cast"].apply(lambda x: [str(actor) for actor in x])
total_data_new["crew"] = total_data_new["crew"].apply(lambda x: [str(crew_member) for crew_member in x])

total_data_new["tags"] = total_data_new["overview"] + total_data_new["genres"] + total_data_new["keywords"] + total_data_new["cast"] + total_data_new["crew"]
```

```python
total_data_new["tags"][0]
```

```python
total_data_new["tags"] = total_data_new["tags"].apply(lambda x: ",".join(x).replace(",", " "))
```

```python
total_data_new["tags"][0]
```

```python
total_data_new.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)

```

```python
total_data_new.iloc[0].tags
```

```python
total_data_new.to_csv("csv_path", index = False)

conn = sqlite3.connect(database_path)

movies.to_sql("clean_movies_data", conn, if_exists = "replace", index = False)
```

```python
# KNN modeling
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(total_data_new["tags"])

model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
model.fit(tfidf_matrix)
```

```python
def get_movie_recommendations(movie_title):
    movie_index = total_data_new[total_data_new["title"] == movie_title].index[0]
    distances, indices = model.kneighbors(tfidf_matrix[movie_index])
    similar_movies = [(total_data_new["title"][i], distances[0][j]) for j, i in enumerate(indices[0])]
    return similar_movies[1:]

input_movie = "How to Train Your Dragon"
recommendations = get_movie_recommendations(input_movie)
print("Film recommendations '{}'".format(input_movie))
for movie, distance in recommendations:
    print("- Film: {}".format(movie))
```
